```python
# imports
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score,accuracy_score,log_loss
import warnings
from sklearn.svm import SVC
warnings.filterwarnings('ignore')
```

```python
sonar_df = pd.read_csv(r'..\Cases\Sonar\Sonar.csv')
sonar_df.head()
```

```python
X= sonar_df.drop(columns=['Class'])
y= sonar_df['Class']
le= LabelEncoder()
y= le.fit_transform(y)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=25,stratify=y)
y_train
```

```python
import numpy as np
scores=[]
c=np.linspace(0.01,5,20)

degree=[2,3,4]
# for i in range(1,19):
for i in c:
    for d in degree: 
        svm=SVC(kernel='linear',C=i,degree=d,probability=True,random_state=25)
        svm.fit(X_train,y_train) 
        y_pred=svm.predict(X_test)
        y_pred_proba=svm.predict_proba(X_test)
        scores.append([i,d,accuracy_score(y_test,y_pred),roc_auc_score(y_test,y_pred_proba[:,1]),log_loss(y_test,y_pred_proba)],)
scores_df=pd.DataFrame(scores,columns=['c','d','Accuracy','ROC_AUC-score','log_loss'])
scores_df.sort_values(by='log_loss',ascending=True,inplace=True)
scores_df.head()
```
