```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
milk=pd.read_csv(r'..\Datasets\monthly-milk-production-pounds-p.csv',index_col=0)
milk.index=pd.DatetimeIndex(milk.index)
milk.head()
```

```python
milk.plot()
plt.title("Monthly milk production")
plt.xlabel("Monthly")
plt.show()
```

```python
downsampled=milk.resample('QE').sum()
downsampled.index.rename('Querter',inplace=True)
downsampled.plot()
plt.title("Quertly milk production")
plt.xlabel("Monthly")
plt.show()
```

### Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
series=milk['Milk']
results=seasonal_decompose(series,model='additive',period=12)
results.plot()
plt.title("Additative Decompose")
plt.show()
```
