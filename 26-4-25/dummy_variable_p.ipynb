```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
from sklearn.linear_model import LinearRegression,Lasso, Ridge
```

```python
df = pd.read_csv(r'C:\Users\ONKAR\OneDrive\Desktop\Machine learning\Machine learning\Datasets\Exp_Salaries.csv')
df.head()
```

```python
train,test = train_test_split(df, test_size=0.3, random_state=25)
x_train = train[['Department']]
y_train = train[['Salary']]
x_test = test[['Department']]
y_test = test[['Salary']]
```

```python
x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)    
```

```python
x_train.head(5)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print("Linear Regression MSE: ", mean_squared_error(y_test, y_pred))
print("Linear Regression Coefficients: ", lr.coef_)
print("Linear Regression Intercept: ", lr.intercept_)   
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel('True Salary')   
plt.ylabel('Predicted Salary')
plt.title('Linear Regression')  
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.show()
```

```python
lasso = Lasso()
lasso.fit(x_train,y_train)
y_pred = lasso.predict(x_test)
```
