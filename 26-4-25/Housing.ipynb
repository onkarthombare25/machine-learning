```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import os
from sklearn.preprocessing import LabelEncoder
```

```python
test_house = pd.read_csv(r'..\Datasets\tst_Housing.csv')
house_df = pd.read_csv(r'..\Datasets\Housing.csv')
test_house.head()
house_df.head()
```

```python
le=LabelEncoder()
house_df['driveway'] = le.fit_transform(house_df['driveway'])
house_df['recroom'] = le.fit_transform(house_df['recroom'])
house_df['fullbase'] = le.fit_transform(house_df['fullbase'])
house_df['gashw'] = le.fit_transform(house_df['gashw'])
house_df['prefarea'] = le.fit_transform(house_df['prefarea'])
house_df['airco'] = le.fit_transform(house_df['airco'])
house_df.head()

```

```python
sc=StandardScaler()
sc.fit_transform(house_df.drop('price',axis=1))
```

```python
x , y = house_df.drop(['price'],axis=1), house_df['price']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=25)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import Ridge
alpha_set=[0.001,0.01,0.1,1,2]
# 0.001 is the alpha value which is the regularization parameter
for i in alpha_set:
    ridge=Ridge(alpha=i)
    ridge.fit(x_train,y_train)
    y_pred = ridge.predict(x_test)
    print("\nalpha",i)
    print("mean_absolute_error",mean_absolute_error(y_test, y_pred))
    print("mean_squared_error",mean_squared_error(y_test, y_pred))
    print("r2_score",r2_score(y_test, y_pred))
```

```python
from sklearn.linear_model import Lasso
alpha_set=[0.001,0.01,0.1,1,2]
# 0.001 is the alpha value which is the regularization parameter
for i in alpha_set:
    ridge=Lasso(alpha=i)
    ridge.fit(x_train,y_train)
    y_pred = ridge.predict(x_test)
    print("\nalpha",i)
    print("mean_absolute_error",mean_absolute_error(y_test, y_pred))
    print("mean_squared_error",mean_squared_error(y_test, y_pred))
    print("r2_score",r2_score(y_test, y_pred))
```

```python
#Elastic net regression
from sklearn.linear_model import ElasticNet

#Testing different alpha values and l1_ratio values to find the best combination.
ration=[0.01,0.25,0.5,0.75,0.9]
alpha_set=[0.001,0.01,0.1,0.5,1,2] 
scores=[]
for i in alpha_set:
    for j in ration:
        el=ElasticNet(alpha=i, l1_ratio=j,random_state=25)
        el.fit(x_train,y_train)
        y_pred = el.predict(x_test)
        scores.append([j,i,r2_score(y_test, y_pred)])
scores=pd.DataFrame(scores, columns=['ratio','alpha','r2_score'])
scores.sort_values(by='r2_score', ascending=False).head(10)
```
