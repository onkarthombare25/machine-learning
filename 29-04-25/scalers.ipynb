```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
df_milk = pd.read_csv(r'..\Datasets\milk.csv',index_col='Animal')
df_milk.head()
```

```python
scaler=StandardScaler().set_output(transform='pandas')
scaler.fit(df_milk)
```

```python
scaled_milk=scaler.transform(df_milk)
scaled_milk.describe()
```

### using minmax scaller


```python
scaler_minmax=MinMaxScaler().set_output(transform='pandas')
scaler_minmax.fit(df_milk)
minmax_scaled_milk=scaler_minmax.transform(df_milk)
minmax_scaled_milk.describe()
```
