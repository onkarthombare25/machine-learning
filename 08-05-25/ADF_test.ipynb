```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import root_mean_squared_error
from statsmodels.graphics.tsaplots import plot_acf
```

```python
fmc=pd.read_csv(r'..\Datasets\FMAC-HPI_24420.csv')
y=fmc['NSA Value']
fmc.head()
```

```python
results=adfuller(y,maxlag=10)
if results[1]<0.05:
    print('Time searies is stationary')
else:
    print('Time searies is NOT stationary')
```

```python
diff1=y.diff()[1:]
results=adfuller(diff1,maxlag=10)
if results[1]<0.05:
    print('Time searies is stationary')
else:
    print('Time searies is NOT stationary')
```

```python
diff2=diff1.diff()[1:]
results=adfuller(diff2,maxlag=10)
if results[1]<0.05:
    print('Time searies is stationary')
else:
    print('Time searies is NOT stationary')
```

```python
plot_acf(y,lags=20,alpha=None)
plt.show()
```
